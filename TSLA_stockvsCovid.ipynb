```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read in our data
equities_df = pd.read_csv("resources/cleaned_close.csv")
```

```python
stock_df = equities_df.loc[equities_df['symbol'] == "TSLA"]
```

```python
stock_df.head()
```

```python
x_axis = stock_df["date"]
y_axis = stock_df["iexClose"]
```

```python
x_axis.head()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='TSLA')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Value")
# Set the y limit between 0 and 4000.
plt.ylim(0, 2500)
# Create a title.
plt.title("TSLA Stock")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='TSLA')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Value")
# Set the y limit between 0 and 4000.
plt.ylim(0, 2500)
# Create a title.
plt.title("TSLA Stock")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create bar plot.
plt.bar(x_axis, y_axis, color="green", label='TSLA')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Value")
# Create a title.
plt.title("TSLA Stock")
# Add the legend.
plt.legend()
```

```python
# Create axis for Covid deaths
z_axis = stock_df["deathIncrease"]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='TSLA')
plt.plot(x_axis, z_axis, marker="*", color="red", linewidth=2, label='Covid Deaths')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Value")
# Set the y limit between 0 and 4000.
plt.ylim(0, 4000)
# Create a title.
plt.title("TSLA Stock vs Covid Deaths")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```
