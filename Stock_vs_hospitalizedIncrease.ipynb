```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read in our data
equities_df = pd.read_csv("resources/post831.csv")
```

```python
stock = str(input("What stock would you like to look at?")).upper()
```

```python
df = equities_df.loc[equities_df['symbol'] == stock]
```

```python
df.head()
```

```python
# Create Xaxis for Date
x_axis = df["date"]
# Create Yaxis for stock price
y_axis = df["iexClose"]
# Create Zaxis for Covid deaths
y2_axis = df["hospitalizedIncrease"]

#Set chart size
plt.rcParams["figure.figsize"] = (20,6)
```

```python
x_axis.head()
```

```python
# Create stock line chart
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label=stock)

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("$Value")
plt.xticks(rotation=45)

# Set the y to autoscale
plt.autoscale(enable=True, axis='y')

# Create a title.
plt.title(stock)
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create stock bar plot.
plt.bar(x_axis, y_axis, color="green", label=stock)

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)

# Set the y to autoscale
plt.autoscale(enable=True, axis='y')

# Create a title.
plt.title(stock)
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="^", color="blue", linewidth=1, label=stock)
plt.plot(x_axis, y2_axis, marker="*", color="red", linewidth=1, label='hospitalizedIncrease')

#Set chart size
# plt.rcParams["figure.figsize"] = (20,6)

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("$Value")
plt.xticks(rotation=45)

# Set the y to autoscale
plt.autoscale(enable=True, axis='y')

# Create a title.
plt.title("Stock vs hospitalizedIncrease")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(x_axis, y_axis, color="blue", marker="^", linewidth=1, label=stock)

# set x-axis label
ax.set_xlabel("Date", fontsize = 20)
plt.xticks(rotation=45)
# set y-axis label
ax.set_ylabel("Stock Price $", color="blue", fontsize=24)
plt.legend(loc = "lower left")

# twin object for two y-axis on the plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(x_axis, y2_axis,color="red",marker="*",linewidth=1, label='Hospitalized Increase')
ax2.set_ylabel("Hospitalized Increase",color="red",fontsize=24)

# Set the y to autoscale
plt.autoscale(enable=True, axis='y')

# Create a title.
title = stock + " vs Hospitalized Increase"
plt.title(title,fontsize=24)

# Add a grid.
plt.grid()
# Add the legend.
plt.legend(loc = "best")



plt.show()
```

```python
# Create File Name
format = ".JPG"
hospital = "_hospitalized_Increase"
SaveAs = stock + hospital + format

# save the plot as a jpeg
fig.savefig(SaveAs, format='jpeg', dpi=100, bbox_inches='tight')
```
